```python
def func(n):
    ans = 0
    i = 2
    while i <= 1e8:
        if n == 1:
            break
        elif n % i == 0:
            ans = ans + 1
            n = n // i
        else:
            i = i + 1
    if n > 1:
        ans = ans + 1
    return ans

n = int(input())
u = func(n)
if u == 20:
    print('Yes')
else:
    print('No')
```

```python
import math
def prime_or_not_prime(x):
    '''
    Definition of simple and non-simple numbers.
    
    :param x: argument
    :type x: int
    
    :return: found simple and non-simple numbers
    :type: bool
    
    '''
    if x < 2:
        return False
    for i in range(2, int(x ** 0.5 + 1)):
        if x % i == 0:
            return False
    else:
        return True


def fact(x):
    '''
    Checking divisibility or indivisibility.
    
    :param x: argument
    :type x: int
    
    :return summ: the found number of divisible numbers
    :type summ: int
    
    '''
    summ = 0
    while x % 2 == 0:
        summ = summ + 1
        x = x // 2
    for j in range(3, int(math.sqrt(x)) + 1, 2):
        while x % j == 0:
            summ = summ + 1
            x = x // j
    if x > 2:
        summ = summ + 1
    return summ


def greatest_common_divisor(x, y):
    '''
    Finding the greatest common divisor.
    
    :param x, y: arguments
    :type x, y: int
    
    :return y: the greatest common divisor
    :type y: int
    
    '''
    if x == 0:
        return y
    return greatest_common_divisor(y % x, x)


n = int(input())
a = math.floor(math.log(n) / math.log(2))
summa = 0
while n % 2 == 0:
    summa = summa + 1
    n = n // 2
y = x = 2 % n
for i in range(1, a + 1):
    y = ((y ** 2) + 2) % n
while (n != 1 and summa < 21):
    x = y = 2 % n
    a = math.floor(math.log(n) / math.log(2))
    for i in range(1, a + 1):
        y = ((y ** 2) + 2) % n
    while (greatest_common_divisor(n, abs(x - y)) == 1 and summa < 21):
        x = ((x ** 2) + 2) % n
    else:
        number = greatest_common_divisor(n, abs(x - y))
        if number > 1907348632813 * 2:
            n = 1
            break
        if prime_or_not_prime(number):
            summa = summa + 1
        else:
            summa = summa + fact(number)
        n = n // number
if summa == 20:
    print('Yes')
else:
    print('No')
```
